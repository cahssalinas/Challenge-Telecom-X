````
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
````

````
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Link direto para o arquivo (usando o RAW do GitHub para não falhar)
url_dados = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/main/TelecomX_Data.json"

try:
    # 2. Lendo os dados direto da nuvem
    df_bruto = pd.read_json(url_dados)

    # 3. Pequeno tratamento rápido (exemplo de limpeza para podermos analisar)
    # Vamos garantir que os nomes das colunas e tipos estejam prontos
    df_analise = df_bruto.copy()

    # Se houver valores vazios em cobranças, vamos preencher com a média
    if 'TotalGeral' in df_analise.columns:
        df_analise['TotalGeral'] = pd.to_numeric(df_analise['TotalGeral'], errors='coerce').fillna(0)

    print("✅ Dados carregados e tratados em memória! Pronto para os gráficos.")

    # --- AGORA OS GRÁFICOS QUE VOCÊ PRECISA PARA O DESAFIO ---

    plt.figure(figsize=(12, 5))

    # Gráfico 1: Proporção de Churn
    plt.subplot(1, 2, 1)
    sns.countplot(x='Churn', data=df_analise, palette='viridis')
    plt.title('Total de Evasão (Churn)')

    # Gráfico 2: Churn por Tipo de Contrato (Onde o problema geralmente está!)
    plt.subplot(1, 2, 2)
    sns.countplot(x='TipoContrato', hue='Churn', data=df_analise, palette='magma')
    plt.title('Evasão por Tipo de Contrato')

    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"❌ Erro crítico: {e}")
````

````
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import requests

# 1. Link do arquivo
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/main/TelecomX_Data.json"

try:
    # 2. Baixando o JSON bruto
    response = requests.get(url)
    dados_brutos = response.json()

    # 3. ACHATANDO OS DADOS (O segredo para o erro sumir!)
    # Isso transforma as sub-colunas em colunas reais
    df_analise = pd.json_normalize(dados_brutos)

    # 4. Ajustando os nomes das colunas (O Pandas coloca pontos, ex: 'conta.contrato')
    # Vamos renomear para os nomes que o Seaborn vai entender
    # Nota: Verifique os nomes reais com df_analise.columns
    colunas_renomear = {
        'conta.contrato': 'TipoContrato',
        'churn': 'Churn',
        'cliente.genero': 'Genero'
    }
    df_analise.rename(columns=colunas_renomear, inplace=True, errors='ignore')

    print("✅ Dados normalizados com sucesso!")
    print(f"Colunas disponíveis agora: {df_analise.columns.tolist()[:10]}...")

    # --- GRÁFICOS CORRIGIDOS ---
    plt.figure(figsize=(14, 5))

    # Gráfico 1: Churn (Com o ajuste da 'palette' para evitar o aviso/warning)
    plt.subplot(1, 2, 1)
    sns.countplot(data=df_analise, x='Churn', hue='Churn', palette='viridis', legend=False)
    plt.title('Total de Evasão (Churn)')

    # Gráfico 2: Churn por Tipo de Contrato
    # Verificamos se a coluna existe antes de plotar para não dar erro
    col_contrato = [c for c in df_analise.columns if 'contrato' in c.lower() or 'TipoContrato' in c][0]

    plt.subplot(1, 2, 2)
    sns.countplot(data=df_analise, x=col_contrato, hue='Churn', palette='magma')
    plt.title('Evasão por Tipo de Contrato')
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"❌ Erro ao processar os dados: {e}")
    print("\n💡 Dica: Verifique se os nomes das colunas no seu JSON são exatamente estes.")
````

````
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import requests

url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/main/TelecomX_Data.json"

try:
    # 1. Extração e Normalização
    response = requests.get(url)
    df_analise = pd.json_normalize(response.json())

    # Convertendo 'account.Charges.Total' para numérico, tratando erros
    df_analise['account.Charges.Total'] = pd.to_numeric(df_analise['account.Charges.Total'], errors='coerce')

    # Preenchendo NaNs em 'account.Charges.Total' (por exemplo, com a média ou 0)
    df_analise['account.Charges.Total'] = df_analise['account.Charges.Total'].fillna(0) # ou df_analise['account.Charges.Total'].mean()

    print("✅ Colunas mapeadas com sucesso! Gerando insights...")

    # Configuração visual
    sns.set_theme(style="whitegrid")
    plt.figure(figsize=(16, 6))

    # --- GRÁFICO 1: CHURN POR TIPO DE CONTRATO ---
    plt.subplot(1, 2, 1)
    sns.countplot(data=df_analise, x='account.Contract', hue='Churn', palette='viridis')
    plt.title('Evasão por Tipo de Contrato', fontsize=14)
    plt.xlabel('Contrato')
    plt.ylabel('Quantidade de Clientes')

    # --- GRÁFICO 2: CHURN POR SERVIÇO DE INTERNET ---
    # Vamos ver se a Fibra Óptica ou DSL influencia na saída
    plt.subplot(1, 2, 2)
    sns.countplot(data=df_analise, x='internet.InternetService', hue='Churn', palette='magma')
    plt.title('Evasão por Serviço de Internet', fontsize=14)
    plt.xlabel('Tipo de Internet')
    plt.ylabel('Quantidade de Clientes')

    plt.tight_layout()
    plt.show()

    # --- ANÁLISE EXTRA: VALOR MENSAL ---
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df_analise, x='Churn', y='account.Charges.Monthly', palette='Set2')
    plt.title('Distribuição de Gastos Mensais vs Churn')
    plt.show()

except Exception as e:
    print(f"❌ Erro ao gerar gráficos: {e}")
````
